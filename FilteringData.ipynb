```r
library(tidyverse)
```

```r
data <- read_csv("data/CovidData.csv")
head(data)
```

```r
filtered_data <- data |> filter(CLASIFFICATION_FINAL <= 3) |>
            select(AGE, SEX, ASTHMA, OBESITY, DIABETES, CARDIOVASCULAR, TOBACCO)
filtered_data

write_csv(filtered_data, "data/FilteredCovidData.csv")

```

```r
filtered_data2 <- filtered_data |> filter(CLASIFFICATION_FINAL == 2)
filtered_data2
```
